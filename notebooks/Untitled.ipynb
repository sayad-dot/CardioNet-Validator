```python
# EDA_dataset1.ipynb

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Load dataset
df = pd.read_csv('../data/dataset1.csv')  # adjust path if needed

df.columns = df.columns.str.strip()

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Basic info
print("\nShape of dataset:", df.shape)
print("\nColumns:")
print(df.columns)
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())

# Statistical summary
print("\nStatistical summary:")
print(df.describe())

# Target column name (replace if different)
target_col = 'heart disease'  # Change if your target column has another name

# Target distribution
print(f"\nValue counts of target column '{target_col}':")
print(df[target_col].value_counts())

sns.countplot(data=df, x=target_col, palette='pastel')
plt.title('Target Class Distribution')
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# Feature distributions (first 6 numeric features)
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in numerical_cols[:6]:
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Optional: Pairplot (commented out by default due to time)
# sns.pairplot(df[numerical_cols[:5] + [target_col]], hue=target_col)
# plt.show()
```
